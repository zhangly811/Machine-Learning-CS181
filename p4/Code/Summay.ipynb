```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
QLearn_stat = []
for eta in [0.2, 0.4, 0.6, 0.8]:
    for discount in [0.2, 0.4, 0.6, 0.8]:
        best = None
        aver = None
        filename = "QLearn_gready0.1_learn"+str(eta)+"_discount" + str(discount) + ".txt"
        with open(filename)as f:
            count = 0
            for line in f:
                read = line.strip().split(':')
                if count == 500:
                    best = int(read[1])
                if count == 501:
                    aver = round(float(read[1]),1)
                count += 1
            
        QLearn_stat.append([eta, discount, best, aver])

QLearn_stat = pd.DataFrame(QLearn_stat, columns=['Leaning Rate', 'Discount', 'best score', 'average score'])
QLearn_plot_best = QLearn_stat.pivot('Leaning Rate', 'Discount', "best score")
QLearn_plot_aver = QLearn_stat.pivot('Leaning Rate', 'Discount', "average score")

sns.set()
sns.set(font_scale=1.5)
fig, axs = plt.subplots(ncols=2,figsize = (20,8))
sns.heatmap(QLearn_plot_best, annot=True, vmin=0, vmax = 2000, cmap="YlGnBu", fmt='g', ax = axs[0])
axs[0].set_title ('Q-Learning: Highest Score (decreading ' + r'$\epsilon$' + ')')
sns.heatmap(QLearn_plot_aver, annot=True, vmin=0, vmax = 100, cmap="YlGnBu", fmt='g', ax = axs[1])
axs[1].set_title ('Q-Learning: Average Score (decreasing ' + r'$\epsilon$' + ')')
fig.savefig('/Users/dilys/Desktop/Harvard_1617_Spring/CS 181/HW/p4/QLearnDecrease.png')
```
